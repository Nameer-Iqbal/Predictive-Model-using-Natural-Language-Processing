```python
import bs4 as bs
import urllib.request
import re
import nltk

scraped_data = urllib.request.urlopen('https://en.wikipedia.org/wiki/Lionel_Messi')
article = scraped_data.read()
parsed_article = bs.BeautifulSoup(article,'lxml')
paragraphs = parsed_article.find_all('p')
article_text = ""
for p in paragraphs:
   article_text += p.text
```

```python
#Data Scraapping
import nltk
flag = nltk.download('stopwords')
if (flag == "False" or flag == False):
  print("failed to downlaod stop words")
else:
  print("downloaded stop words....")

  class PreProcessText:
      def token_words(self, text):
          words = text.split()
          return words

  helper = PreProcessText()
  words = helper.token_words(text=article_text)
```

```python
from gensim import corpora
from gensim.models import LdaModel

# Tokenize and remove stop words
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
processed_text = [[word for word in doc if word.lower() not in stop_words] for doc in [words]]

# Create a dictionary and corpus
dictionary = corpora.Dictionary(processed_text)
corpus = [dictionary.doc2bow(text) for text in processed_text]

# Train LDA model
lda_model = LdaModel(corpus, num_topics=5, id2word=dictionary, passes=10)

# Print topics
topics = lda_model.print_topics(num_words=4)
for topic in topics:
    print(topic)
```
